1. **Crie um sistema de calculadora com funções. Cada operação matemática deve ser tratada em uma função.**

```python
def soma(a,b):
  return a + b

def subtracao(a,b):
  return a - b

def multiplicacao(a,b):
  return a * b

def divisao(a,b):
  if b == 0:
    return 'Não é possível divisão com denominador zero'
  else:
    return a / b

print('Bacharelado em Inteligência Artificial - PIT')
print('CALCULADORA COM DUAS VARIÁVEIS')

while True:
  print('Escolha a operação que deseja executar:')
  op = int(input('1-SOMA \n2-SUBTRAÇÃO \n3-MULTIPLICAÇÃO \n4-DIVISÃO \n0-SAIR \n'))

  if op == 0:
    print('Obrigado por usar! \nBacharelado em Inteligência Artificial - PIT')
    break

  if op < 0 or op > 4:
    print('Digite uma opção válida!')
    continue

  print()
  a = float(input('Digite o primeiro número: '))
  b = float(input('Digite o segundo número: '))


  match op:
    case 1:
      print('Você escolheu a operação de soma')
      print(f'({a}) + ({b}) = {soma(a,b)}')
    case 2:
      print('Você escolheu a operação de subtração')
      print(f'({a}) - ({b}) = {subtracao(a,b)}')
    case 3:
      print('Você escolheu a operação de multiplicação')
      print(f'({a}) * ({b}) = {multiplicacao(a,b)}')
    case 4:
      print('Você escolheu a operação de divisão')
      print(f'({a})"/"({b}) = {divisao(a,b)}')
  print()
```

2. **Faça um programa que receba uma quantidade determinada de números (o usuário que irá determinar quantos números irá digitar). Armazene todos os números digitados em uma lista. O programa deverá:**

* mostrar quantos números pares e ímpares existem na lista;

* mostrar todos os números pares e ímpares digitados.

  **Crie uma função para fazer essa ação.**

```python
def numeros_digitados(lista):
  par = []
  impar = []

  for x in lista:
    if x % 2 == 0:
      par.append(x)
    else:
      impar.append(x)
  return par, impar

i=[]
y=int(input('Digite a quantidade de números que deseja digitar: '))
for x in range(y):
  i.append(int(input(f'Digite o {x+1}º número: ')))

par, impar = numeros_digitados(i)

print()
print(f'A quantidade digitadas de números pares é: {len(par)}')
print(f'Os números pares são: {par}')
print()
print(f'A quantidade digitadas de números ímpares é: {len(impar)}')
print(f'Os números ímpares são: {impar}')
```

3. **Você faz parte do grupo de analista de dados do Instituto Nacional de Meteorologia (INMET). Todos os dias o setor captura os dados da temperatura. São 8 capturas no dia, começando às 6h da manhã e terminando meia noite. Você deve criar um código para armazenar os dados sobre a temperatura e tirar a média. Sobre a média da temperatura há uma análise da seguinte forma:**

* Temperatura menor que 20: baixa

* Temperatura entre 20 e 35: ideal

* Temperatura entre 36 e 50: em alerta

* Temperatura maior que 50: risco

  **A função do seu código deve ser para calcular e analisar a média.**

```python
def analisar_temp(lista):
  media_temp = sum(lista)/len(lista)
  print()
  print(f'A média calculada das oito temperaturas no dia é: {media_temp:.1f} graus')
  if media_temp < 20:
    return 'Baixa'
  elif media_temp < 36:
    return 'Ideal'
  elif media_temp <=50:
    return 'Em Alerta'
  elif media_temp > 50:
    return 'Em Risco'
  else:
    return 'Em erro! Verifique as informações digitadas.'

temp_dia = []
for x in range (8):
  temp_dia.append(float(input(f'Digite a {x+1}º temperatura do dia: ')))

resultado = analisar_temp(temp_dia)

print()
print(f'Ao analisar a média das oito temperaturas realizadas no dia evidencia que o status do clima está: {resultado}')
```

4. **Um químico está fazendo uma análise de pH em alguns produtos para uma empresa farmacêutica. O profissional terá que receber 12 valores do pH de determinados produtos e analisar se são: ácidos, neutros ou básicos. Faça um algoritmo que ajude o químico nisso. A função em Python deve receber a lista dos dados (pH) digitados dos produtos e separar cada pH, em 3 listas, na seguinte situação:**

* pH menor que 7: ácido

* pH igual a 7: neutro

* pH maior que 7 e até 14: básico

```python
def analisar_ph(lista):
  acidos = []
  neutros = []
  basicos = []

  for ph in lista:
    if ph < 7:
      acidos.append(ph)
    elif ph == 7:
      neutros.append(ph)
    elif ph > 7 and ph <=14:
      basicos.append(ph)
    else:
      print('pH inválido!')

  return acidos, neutros, basicos

valores_ph = []
for x in range (12):
  valores_ph.append(float(input(f'Digite o {x+1}º valor do pH: ')))

acidos, neutros, basicos = analisar_ph(valores_ph)

print()
print(f'pH Ácidos: {acidos}')
print(f'pH Neutros: {neutros}')
print(f'pH Básicos: {basicos}')
```
